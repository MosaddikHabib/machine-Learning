```python
df = df.drop(['yr'], axis=1) #dropping the yr column
```

```python
df = df.drop(['id'], axis=1) #dropping the yr column
```

```python
x = df.iloc[:,:-1]
```

```python
x
```

```python
y = df.iloc[:,11]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size=.3, random_state=1)
```

```python
xtrain
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
Rclf = RandomForestClassifier()
```

```python
Rclf.fit(xtrain, ytrain)
```

```python
Rclf.score(xtest, ytest)
```

```python
# prompt: plot confusion matrix

import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

# Predict the labels for the test data
y_pred = Rclf.predict(xtest)

# Calculate the confusion matrix
conf_matrix = confusion_matrix(ytest, y_pred)

# Plot the confusion matrix
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues")
plt.show()
```

```python
# prompt: plot accuracy

import matplotlib.pyplot as plt
# Calculate the accuracy of the model
accuracy = Rclf.score(xtest, ytest)

# Plot the accuracy
plt.bar("Accuracy", accuracy)
plt.xlabel("Metric")
plt.ylabel("Value")
plt.show()
```

```python
# prompt: plot f1

import matplotlib.pyplot as plt
from sklearn.metrics import f1_score

# Calculate the f1 score
f1 = f1_score(ytest, y_pred)

# Plot the f1 score
plt.bar("f1", f1)
plt.xlabel("Metric")
plt.ylabel("Value")
plt.show()
```

```python
# prompt: plot learning and testing

import matplotlib.pyplot as plt
# Calculate the training and testing scores for each number of trees
train_scores, test_scores = [], []
for n_trees in range(10, 101, 10):
    # Create a new random forest classifier with the current number of trees
    clf = RandomForestClassifier(n_estimators=n_trees)

    # Fit the classifier on the training data
    clf.fit(xtrain, ytrain)

    # Calculate the training and testing scores
    train_scores.append(clf.score(xtrain, ytrain))
    test_scores.append(clf.score(xtest, ytest))

# Plot the training and testing scores
plt.plot(range(10, 101, 10), train_scores, label="Training score")
plt.plot(range(10, 101, 10), test_scores, label="Testing score")
plt.xlabel("Number of trees")
plt.ylabel("Score")
plt.legend()
plt.show()
```
